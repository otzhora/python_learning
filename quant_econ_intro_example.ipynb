```python
# exercise 1

def factorial(n): 
    res = 1
    for i in range(2, n):
        res *= i
    return res

print(factorial(6))
```

```python
# exercise 2

def binomia_rv(n, p): 
    U = uniform(0, 1, n)
    return len(U[U<p])
    
binomia_rv(100, 0.1)
```

```python
# exercise 3

# Monte Carlo pi estitmation 

def est_pi():
    n = 10000
    count = 0
    
    for i in range(n):
        x, y = uniform(0, 1, 2) # generate point on surface 
        r = np.sqrt((x - 0.5) ** 2 + (y - 0.5) ** 2) # distance from (0.5, 0.5)
        if(r < 0.5):
            count += 1 # if (x, y) is inside circle w/ r = 0.5
            
    area = count / n # prob that (x, y) is in circle is equal to its area 
    return area * 4 
    
    
est_pi()
```

```python
# exrecise 4

def flip_coin():
    
    last = 1 if uniform() > 0.5 else 0
    curr = None
    count = 1
    
    for i in range(9):
        curr = 1 if uniform() > 0.5 else 0
        
        if(curr == last):
            count += 1
            if(count == 3):
                return "You won a dollar"
        else:
            count = 1
            last = curr
    return "Hah, loser"

flip_coin()
        
        
```

```python
# excercise 5

def get_time_series(alp):
    eps = uniform(-1, 1, 200)
    x = np.zeros(200)

    for i in range(1, 200):
        x[i] = x[i - 1] * alp + eps[i]
    return x

plt.plot(get_time_series(0.9))
plt.show()
```

```python
# excercise 6

plt.plot(get_time_series(0.98), 'r-',label="0,98")
plt.plot(get_time_series(0.9), 'g-', label="0,90")
plt.plot(get_time_series(0), 'b-', label="0,0")

plt.show()
```
